```python
import numpy as np
import pandas as pd

worldcup = pd.read_csv('wc_player_stats.csv',index_col=0)#index_col sets the index to the first column. In this case, player ID

print(worldcup.shape)
print(type(worldcup))
```

```python
worldcup
```

```python
#1 who had the most goals
worldcup.sort_values('goals_scored',ascending = False) 
```

```python
#2 first 10 players who had the lease goals that are a striker or midfielder
least_wc = worldcup[(worldcup.position != 'GK') & (worldcup.position != 'DF')] 
least_wc.sort_values(by = ['goals_scored' ], ascending =True)[:11]
```

```python
#3 which defenders had the most goals?
goalsDF = worldcup[worldcup.position == "DF"]
goalsDF.sort_values('goals_scored',ascending = False)[:5]
```

```python
#4 which club has the most international players
club = worldcup.groupby(["club"]).size()
club.sort_values(ascending = False)[:5]
```

```python
#5 who had the most assists
worldcup.sort_values('assists',ascending = False)[:5]
```

```python
#6 which midfielder had the most assists
worldcup[worldcup.position == 'MF'].sort_values(by = 'assists',ascending = False)[:5]
```

```python
#7 who had the most appearances 
worldcup.sort_values(by = "appearances", ascending = False)[:20]
```

```python
#8 which defender made the most tackles
worldcup2 = pd.read_csv('wc_player_stats2.csv',index_col=0)
wc = worldcup.merge(worldcup2, on='player_id') #joining two tables on the player id column

SF = wc[wc.position == 'DF'].sort_values(by = 'successful_tackles',ascending = False)[:5]
SF[['player_name','position','successful_tackles']]
```

```python
#9 who had the most minutes played
min_played = wc.sort_values(by = 'minutes_played',ascending = False)[:10]
min_played[['player_name','minutes_played',]]
```

```python
#10 who caused the most fouls

fouls_caused = wc.sort_values(by = 'fouls_caused',ascending = False)[:10]
fouls_caused[['player_name','fouls_caused','position']]
```

```python
#11 who committed the most fouls
fouls_caused = wc.sort_values(by = 'fouls_commited',ascending = False)[:10]
fouls_caused[['player_name','fouls_commited','position']]
```

```python
#12 players that caused the most fouls grouped by clubs 
clubs_caused = wc.groupby(['club'])['fouls_caused'].sum()
clubs_caused.sort_values(ascending = False)[:10]
```

```python
#using the aggregate function
clubs = wc.groupby(['club'])['fouls_caused'].agg('sum')
clubs.sort_values(ascending = False)[:10]
```

```python
#13 what national team caused the most fouls
nations_caused = wc.groupby(['nationality'])['fouls_caused'].sum()
nations_caused.sort_values(ascending = False)[:10]
```

```python
#14 players that comitted the most fouls grouped by clubs
club_committed = wc.groupby(['club'])['fouls_commited'].agg('sum')
club_committed.sort_values(ascending = False)[:10]
```

```python
#15 what national team commited the most fouls
nation_comitted = wc.groupby(['nationality'])['fouls_commited'].sum()
nation_comitted.sort_values(ascending = False)[:10]
```

```python
#16 Which player has the most successful dribbles
dribbles = wc.sort_values(['successful_dribbles'],ascending = False)
dribbles[['player_name','successful_dribbles']][:10]
```

```python
#17 Who completed the most passes
passes = wc.sort_values('pass_completed',ascending = False)
passes[['player_name','pass_completed']][:10]
```

```python
# 18 which team had the most shots
shots = wc.groupby(['nationality'])['shots'].sum()
shots.sort_values(ascending = False)[:10]
```

```python
# 19 which team had the most shots on target
sot = wc.groupby(['nationality'])['shots_on_target'].sum()
sot.sort_values(ascending = False)[:10]
```

```python
#20 which age group has the most goals
pd.set_option('display.max_rows', None)
cutoff = [15,20,25,30,35,40]
labels = ['15-20','20-25','25-30','30-35','35-40']
wc['Age Group'] = pd.cut(wc['Age'],bins = cutoff, labels = labels) #function that groups a column and creates another column
wcGroup = wc.groupby(by = 'Age Group')['goals_scored'].sum()
wcGroup.sort_values(ascending = False)
```

```python
#21 which age group has the most assists
wcAssists = wc.groupby(by = 'Age Group')['assists'].sum()
wcAssists.sort_values(ascending = False)
```

```python
#22 which age group has the most minutes on average
wcPlayed = wc.groupby(by = 'Age Group')['minutes_played'].mean().round(2)
wcPlayed.sort_values(ascending = False)
```

```python
#23 which national team had the least tackles
least_tackles = wc.groupby(['nationality'])['successful_tackles'].sum()
least_tackles.sort_values()
```

```python
#24 Which player has the highest pass completed to minutes ratio. Only players that played more than 90 minutes(1 full game)
wcMin = wc[wc.minutes_played > 90]
wcMin['pass to minute ratio'] = (wcMin['pass_completed'] / wcMin['minutes_played']).round(2)
wcMin
```
